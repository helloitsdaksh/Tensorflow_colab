#EXTRA-CIRRCULUM

```python
#Part 1
tf.math.logical_not(tf.constant([True, False]))
```

```python
#Part 2 shape(x,y,z) 
#x = years
#y = months
#z = no.of items
G = tf.constant(np.array([np.random.randint(0,50,(12,5))]),shape = (1,12,5))
G
```

```python
#sum of months
sum_month = tf.reduce_sum(G,axis = 2)
sum_month.numpy()
```

```python
sum_year = tf.reduce_sum(sum_month)
sum_year.numpy()
```
